```python
import pandas as pd
```

# Extra Tree Classifier 

```python
#now training the data
# 1.using extra tree classifier
from sklearn.ensemble import ExtraTreesClassifier
ex=ExtraTreesClassifier()
ex.fit(x,y)
```

```python
#prediction in test dataset
y_pred_ex=ex.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score,recall_score,f1_score,roc_auc_score

print(f'Accuracy score of extra tree classifier = {accuracy_score(y_test,y_pred_ex)}')
print(f'Precision score of extra tree classifier = {precision_score(y_test,y_pred_ex,average='macro')}')
print(f'Recall score of extra tree classifier = {recall_score(y_test,y_pred_ex,average='macro')}')
print(f'F1 score of extra tree classifier = {f1_score(y_test,y_pred_ex,average='macro')}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_ex))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
csvc_matrix = confusion_matrix(y_test,y_pred_ex)
plt.figure(figsize=(12,8))
sns.heatmap(csvc_matrix, annot=False)
plt.title("Confusion Matrix for extra tree classifier")
plt.show()
```
